## 一、if-else结构

```python
# 根据给定的分数进行等级划分。
# 参数:
#     score (float or int): 分数 (通常在 0~100 之间)
# 返回:
#     str: 等级 (如 'A', 'B', 'C', 'D', 'F')

score = 88
# 先检查分数是否在合理范围
if score < 0 or score > 100:
    print("Invalid score")
# 根据分数区间进行定级
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

## 二、for循环

### 1. 遍历

#### 1.1 从小到大打印

```python
start = 0
end = 10
interval = 2
for i in range(start, end, interval):
    print(i)
```

#### 1.2 从大到小打印

```python
start = 0
end = 10
interval = -2
for i in range(end, start, interval):
    print(i)
```

### 2.直角三角形

#### 2.1 正三角

```python
height = 6  # 三角形的高
for i in range(1, height + 1):
    print('*' * (2*i - 1))
```

#### 2.2 倒三角

```python
height = 6  # 三角形的高
for i in range(height, 0, -1):
    print('*' * (2*i - 1))
```

### 3.等腰三角形

#### 3.1 正三角

```python
height = 6  # 三角形的高
for i in range(1, height + 1):
    # 打印空格和*
    for j in range(height - i):
        print(" ", end="")
    for k in range(2 * i - 1):
        print("*", end="")
    print()
```

#### 3.2 倒三角

```python
height = 6  # 三角形的高
for i in range(height, 0, -1):
    # 打印空格和*
    for j in range(height - i):
        print(" ", end="")
    for k in range(2 * i - 1):
        print("*", end="")
    print()
```
